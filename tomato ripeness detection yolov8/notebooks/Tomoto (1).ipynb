```python
from ultralytics import YOLO

# Load pretrained YOLOv8 nano model
model = YOLO("yolov8n.pt")  # 'n' = small & fast
```

```python
model.train(
    data="/content/drive/MyDrive/tomato/data.yaml",  # path to your YAML
    epochs=10,      # fewer epochs to save time
    imgsz=640,      # image size
    batch=8,        # adjust if GPU is small
    name="tomato_ripeness_model"
)
```

```python
!ls /content/runs/detect/tomato_ripeness_model4/weights
```

```python
trained_model = YOLO(
    "/content/runs/detect/tomato_ripeness_model4/weights/best.pt"
)
```

```python
trained_model.predict(
    source="/content/drive/MyDrive/tomato/images/test",
    save=True
)
```
